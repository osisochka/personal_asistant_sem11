 Git — это «машина времени» для твоего кода, которая позволяет вернуться к любой версии проекта, увидеть, кто и когда вносил изменения. Ещё с Git можно работать над новыми функциями параллельно с основным кодом, не боясь что-то в нём сломать.

## Характеристики Git

Git — это программа, которая позволяет:
- увидеть, как выглядел код в определённый момент времени;
- увидеть, кто и когда вносил изменения в код;
- отменять эти изменения при необходимости.

С Git можно работать в графическом или текстовом интерфейсе.

В текстовом интерфейсе используется командная строка. В графическом интерфейсе для работы используются кнопки и меню.  

Большинство современных редакторов кода имеют встроенный Git-клиент, который позволяет работать с Git через графический интерфейс пользователя. Такой вариант проще для новичков, но всё же полезно научиться работать с Git с помощью командной строки. Это поможет:

Перед тем как установить Git, рассмотрим некоторые термины.

### Терминология

#### Репозиторий

> **Репозиторий Git** — это хранилище, в котором расположен проект и его история.

Это может быть локальное хранилище на компьютере или удалённое хранилище на сервисе, например на GitHub. Репозиторий нужен, чтобы отслеживать историю изменений в проекте. Он очень пригодится, если над проектом работает несколько людей и нужно координировать их совместную работу.

> **Директория** — это папка, в которой находятся файлы проекта. Она не хранит никакой информации о версиях этих файлов или об их изменениях.

Если инициализировать репозиторий Git в директории, то Git создаст скрытую поддиректорию под названием `.git`, в которой будет храниться вся информация о репозитории. Эта информация включает историю всех изменений, внесённых в репозиторий, а также его текущее состояние.

Только когда ты инициализируешь репозиторий командой `git init`, Git создаёт скрытую папку `.git` внутри директории. Именно эта папка делает твою директорию полноценным репозиторием, сохраняя всю историю изменений и необходимые данные для отслеживания версий.


#### Коммит

> **Коммит** (от англ. commit) — фиксация изменений в проекте.  

Коммит похож на снимок проекта в определённый момент времени. Коммит содержит информацию только о тех изменениях, которые были внесены в репозиторий с момента последнего коммита. Он не содержит все файлы репозитория (если только это не первый коммит). Таким образом, каждый коммит — это небольшой кусочек истории репозитория, основанный на предыдущем коммите. Все они связаны между собой в цепочку, формируя историю изменений твоего проекта.

#### Ветка

> **Ветка** — это параллельная версия репозитория.

Ветки позволяют работать над отдельными функциями проекта, не влияя на основную версию. Закончив работу над новой фичей, ты можешь объединить эту ветку с основной версией проекта.

### Установка Git

*   **Windows:** перейди по [ссылке](https://git-scm.com/downloads/win) и скачай установщик нужной тебе версии. Запусти его и следуй инструкции.
*   **MacOS:** Git уже установлен по умолчанию.

*   **Linux:** инструкцию по установке под твой дистрибутив можно найти здесь: https://git-scm.com/downloads/linux.



### Настройка Git

При первом использовании необходимо передать Git твоё имя и электронную почту. Эти данные нужны, чтобы подписывать сделанные тобой изменения. Подписи позволяют отслеживать, кто и когда изменил файл.

Напиши в терминале:
```
git config --global user.name 'Your Name'
git config --global user.email 'Your Email'
```

В результате может показаться, что ничего не произошло, так как данные команды ничего не возвращают (на скринах `user.email` замазан).

git-1.png

Но можно убедиться, что конфиг Git действительно заполнился твоей информацией. Используй команду:

```
git config --list --show-origin
```



В результате ты получишь весь твой конфиг Git. Среди строчек будут введённые только что `user.name` и `user.email`.

git-2.png

Продолжи настройку Git и выполни команду:
```
git config --global init.defaultBranch main
```
 Она нужна, чтобы задать имя ветки по умолчанию. После выполнения команды в проектах основная ветка всегда будет называться ```main```.

При обмене файлами между разными операционными системами (например, Windows и Linux) могут возникнуть проблемы с концами строк. В Windows строки заканчиваются символами CR и LF, а в Linux и Mac — только LF.

Git может решить, что строки были изменены, даже если на самом деле изменений не было. Чтобы этого избежать, нужно настроить Git для корректной обработки концов строк в разных системах, используя соответствующие команды.

Для Linux/Mac:

```
git config --global core.autocrlf input

git config --global core.safecrlf warn
```

Для Windows:

```
git config --global core.autocrlf true

git config --global core.safecrlf warn
```

Настройка завершена!

## Работа с Git

Создай рабочую директорию и первый файл через командную строку, или, по-другому, терминал. Терминал можно выбрать на своё усмотрение. Для Windows на выбор есть «Терминал», Windows PowerShell и Git Bash. Для Linux и Mac подойдёт обычный терминал, встроенный в систему.

### Создание первого коммита

Файл создан, но пока не указано, что его нужно внести в репозиторий. Текущее состояние репозитория можно посмотреть с помощью команды:
```
git status
```

На скрине написано, что коммитить пока нечего, но есть один untracked file (неотслеживаемый файл), который можно добавить к пулу изменений с помощью команды:
```
git add
```
Сделай это и посмотри, как изменится статус файла.

git-5.png

Если хочешь проиндексировать сразу все файлы, то добавь флаг `-A` . Это значит All, то есть все изменённые или новые файлы.

Индексация в Git — это процесс добавления изменений в так называемую область подготовленных изменений (англ. staging area). Когда ты изменяешь файлы в проекте, они сначала остаются неотслеживаемыми Git. Чтобы подготовить их для создания коммита, нужно проиндексировать эти изменения.

Это означает, что ты сообщаешь Git, какие файлы или изменения нужно зафиксировать в следующем коммите. Использование команды `git add` с флагом `-A` позволяет сразу добавить все изменённые или новые файлы в эту область для последующего коммита.

Команда:
```
git add -A
```

### История
Получение списка произведённых изменений — функция команды ```git log```.

Ты полностью контролируешь то, что отображает ```log```. Например, можно выводить однострочный формат:
 ```
 git log --pretty=oneline
 ```
В нём не будет информации об авторе и времени изменений, но они далеко не всегда нужны и занимают пару лишних строк на экране.

image.png

Существует огромное количество вариантов просмотра истории, ты можешь найти их на странице руководства [git-log](https://git-scm.com/docs/git-log).

### Ветки

Разработка **фичи** (новой функциональности) может быть рискованной и длительной, поэтому лучше изолировать её в отдельной ветке. Когда фича будет готова, ты сможешь слить эту ветку с `main` (основной веткой). Это защитит `main` от рискованного и непроверенного кода.

Также можно работать над несколькими фичами параллельно, создавая для каждой отдельную ветку. Это позволяет вносить изменения в основную ветку, например для исправления ошибок, без вмешательства в текущие разработки.

Изначально новая ветка будет копией основной, а последующие изменения нужно будет синхронизировать специальными командами, о которых мы поговорим позже.

#### Создание новой ветки

Создай ветку ```new_branch``` в твоём репозитории и посмотри на то, как изменился статус с помощью команды:
```
git switch -c new_branch
```
Команда ```switch``` отвечает за смену ветки. Флаг ```-с``` говорит программе создать ветку. То есть мы создаём ветку и сразу переходим в неё.

#### Слияние веток
### Merge
Чтобы лучше было видно, как работает merge, внеси измения в файл `hello.txt` в `main`-ветке и закоммить.

Видно, что есть две расходящиеся ветки в репозитории. В ветке main для просмотра веток и их расхождения используем команду:
```
git log --all --graph
```


*   Опция ```--all``` гарантирует, что мы видим все ветки, так как по умолчанию в логе показывается только текущая ветка.
*   Опция ```--graph``` добавляет простое дерево коммитов (то есть как коммиты связаны между собой), представленное в виде простых текстовых линий.



image.png

Мы видим обе ветки (```new_branch``` и ```main```), причём ветка ```main``` помечена как ```HEAD```, что означает, что она является текущей.

Слияние (merge) переносит изменения из двух веток в одну. Давай сольём ```main``` с ```new_branch``` c помощью команды:
```
git merge new_branch
```
Посмотри на дерево ещё раз.

image.png

Наши ветки успешно слились. Теперь в ветке `main` появились все изменения из `new_branch`. Путём периодического слияния ветки `main` с веткой `new_branch` ты можешь переносить из `main` любые изменения и поддерживать совместимость изменений `new_branch` с изменениями в основной ветке.

## Заключение

- Git — это система управления версиями, которая позволяет отслеживать изменения в коде, возвращаться к предыдущим версиям и управлять несколькими версиями проекта.
- Репозиторий — это хранилище, в котором содержится проект и его история. Он может быть локальным или удалённым.
- Коммит — это фиксированное состояние проекта в определённый момент времени. Коммиты формируют историю изменений проекта.
- Ветка — параллельная версия репозитория, в которой можно работать над новыми функциями без вмешательства в основную версию.
- Git можно установить и настроить на различных операционных системах, что позволяет гибко работать с проектами в любой среде.
- Репозиторий Git позволяет хранить файлы проекта, отслеживать изменения и сохранять их с помощью коммитов.
- Ветки в Git позволяют параллельно работать над разными функциями проекта и легко интегрировать изменения в главную ветку.
- Основные команды Git обеспечивают эффективное управление проектами и их версиями, помогая разработчикам легко контролировать и координировать изменения:
 - `git init` — инициализация нового Git-репозитория;
 - `git config` — настройка параметров Git, таких как имя пользователя и адрес электронной почты;
 - `git add` — добавление файлов в индекс для последующего коммита;
 - `git commit` — создание коммита для сохранения проиндексированных изменений;
 - `git status` — проверка текущего статуса репозитория, отображение изменений и неотслеживаемых файлов;
 - `git log` — просмотр истории коммитов в репозитории;
 - `git switch` — переключение между ветками и создание новой ветки;
 - `git merge` — слияние изменений из одной ветки в другую;
 - `git log --graph` — отображение истории коммитов в виде графа, что позволяет визуализировать структуру веток.
